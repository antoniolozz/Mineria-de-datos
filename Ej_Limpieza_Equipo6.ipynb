#### Investigar los siguientes comandos: 
* ?df.dropna()

El método dropna permite, de una forma muy conveniente, filtrar los valores de una estructura de datos para dejar solo aquellos no nulos.

* ?df.astype

El método se usa para convertir un objeto pandas a un tipo de dtype específico. La función también proporciona la capacidad de convertir cualquier columna existente adecuada a un tipo categórico.
